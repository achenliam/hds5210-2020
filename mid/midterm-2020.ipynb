Write a series of functions, enough to satisfy all of the main criteria that we're using to calculate the Apache II score.  That list is the same as the assert statements above.

````python
import json
def Organ_Failure_History(Organ_F_Hist, config_file):
    """str, str -> int
    Organ_F_Hist: History of organ failure
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    OFH_rules = config.get("Organ Failure History")
    points = OFH_rules[Organ_F_Hist]
    return points
def Age_score(ages, config_file):
    """int, str -> int
    ages: The age of patient
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    age_rules = config.get("Age")
    for rules in age_rules:
        if ages >= rules.get('min') and ages <= rules.get('max'):
            points = rules.get('points')
    return points
def Temperature_score(temp, config_file):
    """float, str -> int
    temp: The body temperature of the patient 
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    
    """
    points = 0
    config = json.load(open(config_file))
    temp_rules = config.get("Temperature")
    for rules in temp_rules:
        if temp >= rules.get('min') and temp < rules.get('max'):
            points = rules.get('points')
    return points
def pH(ph, config_file):
    """float, str -> int
    ph: Arterial pH of the patient
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    ph_rules = config.get("pH")
    for rules in ph_rules:
        if ph >= rules.get('min') and ph < rules.get('max'):
            points = rules.get('points')
    return points
def Heart_rate(hr, config_file):
    """int, str -> int
    hr: The heart beat rate of the patient (beat/min)
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    hr_rules = config.get("Heart rate")
    for rules in hr_rules:
        if hr >= rules.get('min') and hr < rules.get('max'):
            points = rules.get('points')
    return points
def Respiratory_rate(rr, config_file):
    """int, str -> int
    rr: The rate of respiration of patient (breath/min)
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    rr_rules = config.get("Respiratory rate")
    for rules in rr_rules:
        if rr >= rules.get('min') and rr < rules.get('max'):
            points = rules.get('points')
    return points
def Sodium(Na, config_file):
    """int, str -> int
    Na: The level of sodium in patient's blood sample (mmol/L)
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    Na_rules = config.get("Sodium")
    for rules in Na_rules:
        if Na >= rules.get('min') and Na < rules.get('max'):
            points = rules.get('points')
    return points
def Potassium(K, config_file):
    """int, str -> int
    Na: The level of potassium in patient's blood sample (mmol/L)
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    K_rules = config.get("Potassium")
    for rules in K_rules:
        if K >= rules.get('min') and K < rules.get('max'):
            points = rules.get('points')
    return points
def Creatinine(C, Renewal_Failure,config_file):
    """int, str, str -> int
    C: The level of creatinine in patient's blood sample (mg/100ml)
    Renewal_failure: Renal failure of the patient
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    C_rules = config.get("Creatinine")
    Acute_rules = config.get("Acute")
    Chronic_rules = config.get("Chronics")
    if C < 1.5:
        for rules in C_rules:
            if C >= rules.get('min') and C < rules.get('max'):
                points = rules.get('points')
        return points
    elif C >= 1.5 and Renewal_Failure == "Acute Renal Failure":
        for rules in Acute_rules:
            if C >= rules.get('min') and C < rules.get('max'):
                points = rules.get('points')
        return points
    elif C >= 1.5 and Renewal_Failure == "Chronic Renal Failure":
        for rules in Chronic_rules:
            if C >= rules.get('min') and C < rules.get('max'):
                points = rules.get('points')
        return points
def Hematocrit(H, config_file):
    """int, str -> int
    H: The percetage of hematocrit in patient's blood sample (%)
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    H_rules = config.get("Hematocrit")
    for rules in H_rules:
        if H >= rules.get('min') and H < rules.get('max'):
            points = rules.get('points')
    return points
def White_Blood_Count(WBC, config_file):
    """int, str -> int
    Na: The counts of white blood cell in patient's blood sample (total/cubic mm in 1000's)
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    WBC_rules = config.get("White Blood Count")
    for rules in WBC_rules:
        if WBC >= rules.get('min') and WBC < rules.get('max'):
            points = rules.get('points')
    return points
def FiO2(F, PaO2, gradient, config_file):
    """int, int, int -> int
    F: The percentage of FiO2 in patient's blood sample (%)
    PaO2: The level of PaO2 in patient's blood sample (mmHg)
    gradient: The level of  A-a gradient in patient's blood sample
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    config = json.load(open(config_file))
    greater50_rules = config.get("A-a gradient")
    less50_rules = config.get("PaO2")
    if F <= 50:
        for rules in less50_rules:
            if PaO2 >= rules.get('min') and PaO2 <= rules.get('max'):
                points = rules.get('points')
        return points
    elif F > 50:
        for rules in greater50_rules:
            if gradient >= rules.get('min') and gradient <= rules.get('max'):
                points = rules.get('points')
        return points
def Glasgow_Coma_Scale(GCS):
    """int, str -> int
    GCS: Glasgow coma scale
    config_file: The name of the configuration file to use
    
    The function will return a score for Apache score calculation base on configuration file.
    """
    points = 0
    if GCS <= 15:
        points = 15 - GCS
    return points
````

## Part 3: Put it all together

Create a new function called `apache_score()` that takes all of the necessary inputs and returns the final Apache II score.  Use any variable names that you want.  For clarity and organization, my recommendation is to create them in the same order as they're documented in the website.

````python
def apache_score(Organ_Failure_History, Age_score, Temperature_score, pH, Heart_rate, Respiratory_rate, Sodium, Potassium, Creatinine, Hematocrit, White_Blood_Count, Glasgow_Coma_Scale, FiO2):
    score = 0
    config_file = "apache.json"
    score = Organ_Failure_History(Organ_F_Hist, config_file)
    score = Age_score(ages, config_file) + score
    score = Temperature_score(temp, config_file) + score
    score = pH(ph, config_file) + score
    score = Heart_rate(hr, config_file) + score
    score = Respiratory_rate(rr, config_file) + score
    score = Sodium(Na, config_file) + score
    score = Potassium(K, config_file) + score
    score = Creatinine(C, Renewal_Failure,config_file) + score
    score = Hematocrit(H, config_file) + score
    score = White_Blood_Count(WBC, config_file) + score
    score = Glasgow_Coma_Scale(GCS) + score
    score = FiO2(F, PaO2, gradient, config_file) + score
    return score
````
